#### SQL Server 2012 connection to python using SQL Server Driver

```python
cnxn_str = ("Driver={SQL Server};"
            "Server=DESKTOP-RH4VB6T\SQLEXPRESS;"
            "Database=vamsi;"
            "Trusted_Connection=yes;")
```

```python
connection = pyodbc.connect(cnxn_str)
```

```python
cursor = connection.cursor()
for i in cursor.execute("SELECT  * FROM bus"):
    print(i)
```

```python
cursor.close()
connection.close()
```

#### SQL Server 2012 connection to python using SQL Server Native Client 11.0 Driver

```python
cnxn_str = ("Driver={SQL Server Native Client 11.0};"
            "Server=DESKTOP-RH4VB6T\SQLEXPRESS;"
            "Database=vamsi;"
            "Trusted_Connection=yes;")
cnxn = pyodbc.connect(cnxn_str)
```

```python
cursor = cnxn.cursor()
for i in cursor.execute("SELECT  * FROM bus"):
    print(i)
```

```python
cursor.close()
cnxn.close()
```
